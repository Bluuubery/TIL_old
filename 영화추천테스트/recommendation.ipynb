```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating_data = pd.read_csv('./data/ratings.csv')
movie_data = pd.read_csv('./data/movies.csv')
```

```python
rating_data.drop('timestamp', axis = 1, inplace=True)
```

```python
user_movie_rating = pd.merge(rating_data, movie_data, on='movieId')
```

```python
movie_user_rating = user_movie_rating.pivot_table('rating', index='title', columns='userId')
```

```python
movie_user_rating.fillna(0, inplace=True)
```

```python
item_based_collaborate = cosine_similarity(movie_user_rating)
```

```python
item_based_collaborate = pd.DataFrame(data=item_based_collaborate, index=movie_user_rating.index, columns=movie_user_rating.index)
```

```python
item_based_collaborate.head()
```

```python
def recommend_movie(title):
    return item_based_collaborate[title].sort_values(ascending=False)[:6]
```

```python
recommend_movie('Taxi Driver (1976)')
```
